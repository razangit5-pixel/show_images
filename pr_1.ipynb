```python
# 0 for greyscale
# 1 for color
# -1 alpha channel
img = cv2.imread('apple.jpg',0)
```

```python
img
```

```python
cv2.imshow('apple.jpg',img)
# wait for any key
k = cv2.waitKey(0)
```

```python
cv2.imshow('apple.jpg',img)
# wait for any key
k = cv2.waitKey(0)
# 27 for esc
if k == 27:
    cv2.destroyAllWindows()
# 115 for s
elif k == ord('s'):
    cv2.imwrite('apple_copy.jpg',img)
    cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

img_gray = cv2.imread('apple.jpg', cv2.IMREAD_GRAYSCALE)  #   

plt.imshow(img_gray, cmap='gray')  #  cmap='gray' 
plt.axis('off')
plt.show()

```
